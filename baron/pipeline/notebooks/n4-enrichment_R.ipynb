```r
# Loop through libraries
for (gene_set_library in head(unique(enrichment_dataframe$gene_set_library), 2)) {
    
    # Print
    print(gene_set_library)
    
    # Loop through directions
    for (direction in unique(enrichment_dataframe$direction)) {
        
        # Get plot dataframe
        plot_dataframe <- enrichment_dataframe[(enrichment_dataframe$gene_set_library == gene_set_library) & (enrichment_dataframe$direction == direction),]
        plot_dataframe <- head(plot_dataframe)
        print(plot_dataframe)

        # Plot
        gp <- ggplot(plot_dataframe, aes(x=term_name, y=log10P)) +
            geom_bar(stat='identity') +
            coord_flip() +
            labs(x='asd', 'y'='', title=direction)
        print(gp)
    }
    
}
```
